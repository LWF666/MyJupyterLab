```python
# 学习turtle模块
```

```python
# 导入turtle
import turtle
```

```python
# 显示turtle箭头
turtle.showturtle()
```

```python
# 画出字符串
turtle.write("LWF")
```

```python
# 前进300像素
turtle.forward(300)
```

```python
# 画笔颜色改为red
turtle.color("red")
```

```python
# 箭头左转90度
turtle.left(90)
```

```python
# 去坐标
turtle.goto(0, 50)
```

```python
# 不想划线，抬笔操作
turtle.penup()
```

```python
# 落笔操作
turtle.pendown()
```

```python
# 画圆
turtle.circle(100)
```
